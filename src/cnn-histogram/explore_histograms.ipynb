```python
def plot_pairwise_gene_histogram(idx, idx_sample, save_dir, gene_conn_list, histograms):
    '''
    Inputs:
    idx: index of interest
    save_dir: directory where to save the image
    gene_conn_list: dataframe describing the interaction between genes
    '''

    plt.figure()
    plt.imshow(histograms[idx, :, :, 0])
    genea = gene_conn_list.iloc[int(idx + (idx_sample * len(gene_conn_list)/n_samples))]['GeneA']
    geneb = gene_conn_list.iloc[int(idx + (idx_sample * len(gene_conn_list)/n_samples))]['GeneB']
    label = gene_conn_list.iloc[int(idx + (idx_sample * len(gene_conn_list)/n_samples))]['interaction']
    plt.title(f'{genea} and {geneb} - Label {label}')
    plt.colorbar()
    idx_str = str(idx + (int(idx_sample * len(gene_conn_list)/n_samples)))
    plt.savefig(save_dir / f'histogram_{genea}_{geneb}.pdf')
    return label
```

```python
save_dir = Path('/output/TransformedData_DerivedData/CNN/histogram/histograms')
save_dir.mkdir(exist_ok=True)

n_samples = 4 # how many times the data was splitted
idx = 1
idx_sample = 0
label = plot_pairwise_gene_histogram(idx, idx_sample, save_dir, gene_conn_list, histograms)
print(f'Label {label}')
```
